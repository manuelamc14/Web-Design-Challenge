```python
#Dependencies and Setup

import pandas as pd

#File to load

file_to_load = "cities.csv"

# Read file and store into Pandas data frame
cities_data = pd.read_csv(file_to_load)
# print(cities_data.to_string(index=False))
```

```python
# Cities data frame

cities_data
```
